# pct_distribution.csv

```python
# Load data
df_pct = pd.read_csv('pct_distribution.csv')
```

#### 1.	Using the data ‘pct_distribution.csv’, add a column called ‘% Sales Contribution’ (rounded to 2 decimals) which shows how much percent each product has contributed towards the entire sales (ex. 17.94% of the total sales resulted from Product A).

```python
# Calculate '% Sales Contribution'
df_pct['% Sales Contribution'] = round((df_pct['Sales_dollar'] / df_pct['Sales_dollar'].sum()) * 100, 2)
```

```python
# Check df_pct
df_pct
```

#### a.	Using the Expected result from question #1, sort the column ‘% Sales Contribution’ in a descending order. 

```python
# Sort values by '% Sales Contribution'
df_pct.sort_values(by='% Sales Contribution', ascending=False, inplace=True)
```

```python
# Check df_pct
df_pct
```

#### b.	As seen from the Expected result in above 1a, row indexes are out of order now. Reorganize the indexes so it’s from 0 to 5 again. Everything else should stay the same. 

```python
# Reset index
df_pct.reset_index(inplace=True, drop=True)
```

```python
# Check df_pct
df_pct
```

#### c.	Lastly, let’s put a (dollar sign) infront of the numbers in ‘Sales_dollar’ column, then rename this column to ‘Sales’. There are multiple ways to put a (dollar sign) infront of the ‘Sales’ column, try to do it without importing the package. 

```python
# Rename 'Sales_dollar' column to 'Sales'
df_pct.rename({'Sales_dollar':'Sales'}, axis=1, inplace=True)
```

```python
# Check df_pct
df_pct
```

```python
# Add $ infront of the 'Sales column'
df_pct['Sales'] = '$' + df_pct['Sales'].astype('str')
```

```python
# Check df_pct
df_pct
```

# groupby_practice.csv

```python
# Load data
df_grpby = pd.read_csv('groupby_practice.csv')
```

```python
# Make sure data is loaded properly
df_grpby
```

#### 2.	Using the data ‘groupby_practice.csv’, find out how much each product totaled in sales, then name this column as ‘Sales_dollar_product’ 

```python
# Groupby, sum, then rename
df_grpby.groupby('Product').agg({'Sales_dollar':'sum'}).rename({'Sales_dollar':'Sales_dollar_product'}, axis=1)
```

#### a.	Find out how much each person totaled in sales, then name this column as ‘Sales_dollar_person’

```python
# Groupby, sum, then rename
df_grpby.groupby('Sold_by').agg({'Sales_dollar':'sum'}).rename({'Sales_dollar':'Sales_dollar_person'}, axis=1)
```

#### b.	Find out how much each person totaled in sales per product, then name this column as ‘Sales_dollar_person_product’

```python
# Groupby, sum, then rename
df_grpby.groupby(['Sold_by', 'Product']).agg({'Sales_dollar':'sum'}).rename({'Sales_dollar':'Sales_dollar_person_product'}, axis=1)
```

#### c.	It turns out that your stakeholders don’t want any sales above 50.00 dollars to be in the calculation. Using the data ‘groupby_practice.csv’, perform 2a and 2b again without the sales over 50.00 dollars.

```python
# Select data with Sales_dollar < 50
df_dummy = df_grpby[df_grpby['Sales_dollar'] < 50]
```

```python
df_dummy
```

```python
# Groupby on 1 level
df_dummy.groupby('Sold_by').agg({'Sales_dollar':'sum'}).rename({'Sales_dollar':'Sales_dollar_person'}, axis=1)
```

```python
# Groupby on 2 levels
df_dummy.groupby(['Sold_by', 'Product']).agg({'Sales_dollar':'sum'}).rename({'Sales_dollar':'Sales_dollar_person_product'}, axis=1)
```

#### 3.	Using the data ‘merge_practice.csv’, merge it with the data from question 1 (‘pct_distribution.csv’) to result in data as shown below:

```python
# Merge df_pct and df_merge using 'inner'
df_pct.merge(df_merge, how='inner', on='Product')
```

#### a.	As you might have noticed, after the merge, Product E and F are missing as previously as shown from Question 1 (‘pct_distribution.csv’). Merge it in the way that it still shows those products even though they are not present in the ‘merge_practice.csv’ file.

```python
# Merge df_pct and df_merge using 'left'
df_pct.merge(df_merge, how='left', on='Product')
```

#### b.	Lastly, merge it in the way so that it shows Product G and H from Question 3 (‘merge_practice.csv’) as well.

```python
# Merge df_pct and df_merge using 'outer'
df_pct.merge(df_merge, how='outer', on='Product')
```
